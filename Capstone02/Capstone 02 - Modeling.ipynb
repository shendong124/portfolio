4. Conclusion on the regression and classifier model for rating prediction

All the regression and classifier model we used here to predict the ratings did not perform well. The dataset appears to be unsuitable for predicting ratings through regression or classification models. Several factors contribute to this limitation:

- Imbalanced Rating Distribution:

If certain ratings dominate (e.g., most recipes are rated 4 or 5), Random Forest will be biased towards predicting the majority class. This is especially problematic for regression when the distribution of ratings is skewed.


- Subjectivity of Ratings:

The ratings are likely influenced by subjective factors such as the users' individual tastes or the recipe authors' cooking skills. These aspects are not captured in the features we generated, making it challenging to establish a reliable predictive model.

- Lack of Correlation with Nutrient Content:

As shown in the correlation matrix during the exploratory data analysis (EDA), nutrient content does not exhibit a strong relationship with the ratings. This is likely because the dataset originates from a general-purpose recipe database, not one tailored to a specific audience (e.g., those seeking low-carb or low-fat recipes). The absence of a targeted user base may explain the lack of correlation.

- Next Steps:

Given these challenges, the next logical step is to explore unsupervised learning. By applying clustering algorithms, we can group recipes based on similarities in ingredients, nutrient content, and other features. This approach could have several applications such as recommending recipes from clusters that align with a user's preferences or past choices or Identifying clusters that meet specific dietary requirements (e.g., low-carb, gluten-free, vegan) and using these to make targeted recommendations.

This shift to unsupervised modeling may uncover meaningful patterns in the data that can better address user needs.

1. K-means

```python
#scale the whole dataset to apply K-means in order to identify recipe clusters 
# Identify numerical columns that we want to scale
numerical_columns = ['Calories', 'FatContent', 'SaturatedFatContent', 'CholesterolContent', 'SodiumContent', 'CarbohydrateContent', 'FiberContent', 'SugarContent', 'ProteinContent',
       'TotalTimeMinutes', 'description_length', 'rating_count', 'Instruction_length', 'ingredient_count', 'average_rating']

# Separate the numerical column from the one-hot encoded columns
df_nonscale_kmean = df.drop(columns = numerical_columns)

# Initialize StandardScaler
scaler = StandardScaler()

# Scale numerical columns
num_scale_kmean = scaler.fit_transform(df[numerical_columns])

# Convert scaled data back to DataFrame
df_num_scaled_kmean = pd.DataFrame(num_scale_kmean, columns=numerical_columns, index=df.index)

# Concatenate scaled numerical and one-hot encoded columns
df_scaled_kmean = pd.concat([df_num_scaled_kmean, df_nonscale_kmean], axis=1)

df_scaled_kmean.head()
```

```python
# Convert all column names to strings
df_scaled_kmean.columns = df_scaled_kmean.columns.astype(str)
```

```python
#Create a function to calculate the inertia for a range of cluster numbers, which can be used with the elbow method to identify the optimal number of clusters for the model.
def elbow_method(X):
    inertia = []
    k_values = range(1, 11)
    for k in k_values:
        kmeans = KMeans(n_clusters=k, init='k-means++', random_state=42)
        kmeans.fit(X)
        inertia.append(kmeans.inertia_)
    
    plt.figure(figsize=(8, 5))
    plt.plot(k_values, inertia, marker='o')
    plt.xlabel('Number of Clusters (k)')
    plt.ylabel('Inertia')
    plt.title('Elbow Method for Optimal k')
    plt.show()
```

```python
from sklearn.cluster import KMeans
```

```python
elbow_method(df_scaled_kmean)
```

**With the elbow method analysis we can conclude that the optimal is 4 or 5 clusters**

```python
#run k-means with n_clusters = 5

from sklearn.cluster import KMeans
k = 5

# Fit the KMeans model
kmeans = KMeans(n_clusters=k, init='random', random_state=42)
kmeans.fit(df_scaled_kmean)

# Get the cluster centers and labels
cluster_centers = kmeans.cluster_centers_
labels = kmeans.labels_

# Add cluster labels to the DataFrame fhor each sample
df_scaled_kmean['cluster'] = labels
df_scaled_kmean['cluster'].head()
```

- Cluster 0: High-protein recipes featuring ingredients like meat and poultry, with moderate preparation times.
